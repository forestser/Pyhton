## **1. 데이터 불러오기**
+ 사이킷런에 내장된 이 데이터는 딕션너리 형태
  + data : 피처의 데이터 세트 , numpy 배열
  + target : 분류 시 레이블 값, 회귀일 때는 숫자 결과값 데이터 Set
  + target_names : 개별 레이블의 이름 
  + feature_names : 피처의 이름 
  + DESCR : 데이터 세트에 대한 설명과 각 피처의 설명 

```python
#사이킷런의 보스톤 주택 데이터셋 불러오기 
from sklearn.datasets import load_boston

boston = load_boston()
```




*   CRIM: 지역별 범죄 발생률
*   ZN: 25,000평방피트를 초과하는 거주 지역의 비율
*   NDUS: 비상업 지역 넓이 비율
*   CHAS: 찰스강에 대한 더미 변수(강의 경계에 위치한 경우는 1, 아니면 0)
*   NOX: 일산화질소 농도
*   RM: 거주할 수 있는 방 개수
*   AGE: 1940년 이전에 건축된 소유 주택의 비율
*   DIS: 5개 주요 고용센터까지의 가중 거리
*   RAD: 고속도로 접근 용이도
*   TAX: 10,000달러당 재산세율
*   PTRATIO: 지역의 교사와 학생 수 비율
*   B: 지역의 흑인 거주 비율
*   LSTAT: 하위 계층의 비율
*   MEDV: 본인 소유의 주택 가격(중앙값)

### **EDA(탐색적 데이터분석)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#판다스 데이터프레임
df = pd.DataFrame(boston['data']) 
df.columns = boston['feature_names']
```

```python
df.head()
```

```python
#target 열 추가
df['target'] = boston['target']
df.head()
```

```python
sns.heatmap(df.corr())
plt.show()
```

## **2.train data와 test data 분리**
+ train_test_split() : 테스트 셋과 트레이닝 셋을 나눠주는 함수
  + test_size	test셋 비율
    + 기본  75%, 25% 비율
  + train_size	train셋 비율(default: 테스트셋의 나머지)
  + random_state	랜덤시드번호
    + random_state 옵션에 0을 주면 항상 같은 값이 나오도록 데이터들을 셔플
  + suffle	데이터섞기(default: true)
  + statify	라벨 비율 유지

```python
#테스트 셋과 트레이닝 셋을 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.loc[:,['RM', 'LSTAT']], df['target'], random_state=0 )
```

```python
df.loc[:,['RM', 'LSTAT']]
```

## **3.모델링 : scikit-learn 활용**
+ https://github.com/amueller/odscon-2015
+ 사이킷런 제공 api
![scikit learn cheat sheet](https://1.bp.blogspot.com/-ME24ePzpzIM/UQLWTwurfXI/AAAAAAAAANw/W3EETIroA80/s1600/drop_shadows_background.png)

+ 1)모델 추가


```python
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

+ 2) 파라미터를 결정하고 모델 객체 생성

```python
model_dic = {
    "Linear" : LinearRegression(),
    "SVM" : svm.SVR(),
    "RandomForest" : RandomForestRegressor(),
    "xgboost" : XGBRegressor() 
}
```

+ 3) train 데이터셋을 이용하여 훈련(training) 및 예측(predict)

```python
lg = LinearRegression()
lg.fit(X_train, y_train)
```

```python
model_p = {}

for key in model_dic.keys() :
  print(f'{key} : training')
  model_dic[key].fit(X_train, y_train) 

  print(f'{key} : predict')
  model_p[key] =  model_dic[key].predict(X_test)
```

```python
#결과를 데이터프레임으로 만들기  
model_pdf = pd.DataFrame(model_p)
model_pdf['y_test'] = list(y_test)
model_pdf
```

```python
def gh(col) :
  plt.plot(model_pdf.index, model_pdf['y_test'], 'r-')
  plt.plot(model_pdf.index, model_pdf[col], 'b--')
  plt.title(col)

  plt.show()
```

```python
for col in model_pdf.columns :
  gh(col)
```

## **4.모델평가 : scikit-learn 활용**
+ 회귀의 성능 평가 지표: 대부분 실제 값과 예측값의 오차 평균값에 기반한 평가
+ MAE (Mean Absolute Error) : 실제 값과 예측값의 차이를 절댓값으로 변환해 평균한 것
+ MSE (Mean Squared Log Error) : 실제 값과 예측값의 차이를 제곱해 평균한 것
+ RMSE (Root Mean Squared Eerror) : MSE 같은 오류의 제곱을 구할때 실제 오류 평균보다 더 커지는 특성이 있으므로 MSE에 루트를 씌운 것, 값이 작을 수록 모델의 성능이 좋다
+ R 제곱 : 실제 값의 분싼 대비 예측값의 분산 비율을 지표로 하며, 1에 가까울 수록 예측 정확도가 높다


```python
from sklearn.metrics import mean_squared_error , r2_score
import numpy as np
```

```python
model_rmse = {} 
model_r2 = {} 

for key in model_dic.keys() :
  mse = mean_squared_error(y_test, model_p[key])
  rmse = np.sqrt(mse)

  model_rmse[key] = rmse
  model_r2[key] = r2_score(y_test, model_p[key])
```
